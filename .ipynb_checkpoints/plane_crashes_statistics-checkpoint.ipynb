```python
import pandas as pd
from datetime import datetime
import re as re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
na_values = ["NaN","NA"]
df = pd.read_csv('Data\Airplane_Crashes_and_Fatalities_Since_1908.csv',index_col=None,na_values=na_values)
```

```python
df.head(50)
```

```python
df.dtypes
```

Since the Time column has NaN values i will fill those with 00:00

```python
df['Time'] = df['Time'].fillna('00:00')
```

Finding all strings that dont match the pattern 00:00

```python
for item in df['Time']:
    if not re.match(r'^\d{2}:\d{2}$', item):
        print(item)
```

Function for setting time strings not in format 00:00 to 00:00

```python
def convert_to_00_00_format(time_str):
    
    pattern = r'^\d{2}:\d{2}$'
    
    if re.match(pattern, time_str):
        return time_str  
    else:
        return "00:00"
    
```

```python
df['Time'] = df['Time'].apply(convert_to_00_00_format)
```

Now nothing should be printet.

```python
for item in df['Time']:
    if not re.match(r'^\d{2}:\d{2}$', item):
        print(item)
```

Making new column called DateTime that is a combination of date and time, so i can later convert it to a DateTime object using pandas function to_datetime

```python
df['DateTime'] = df['Date'] + ' ' + df['Time']
```

```python
df.head()
```

Converting all the DateTime strings to DateTime objects using pandas function to_dateTime.

```python
df['DateTime'] = pd.to_datetime(df['DateTime'])
```

```python
df.dtypes
```

Making a list of time gone in seconds since midnight(00:00)

```python
time = df['DateTime'].dt.hour*3600 + df['DateTime'].dt.minute * 60 + df['DateTime'].dt.second
```

```python
time
```

```python
df['Time_Since_Midnight_In_Seconds'] = time
```

```python
df.head()
```

Since all the zero values in the Time_Since_Midnight column was originally undefined i will return them to NaN values, so i can find the average time.

```python
df['Time_Since_Midnight_In_Seconds'].replace(0, pd.NA, inplace=True)
```

```python
df.head()
```

Calculating the mean and filling all NA values with it

```python
time_mean = df['Time_Since_Midnight_In_Seconds'].mean()
```

```python
df['Time_Since_Midnight_In_Seconds'] = df['Time_Since_Midnight_In_Seconds'].fillna(time_mean)
```

```python
year_count = df.groupby(df.DateTime.dt.year)[['Date']].count()

plt.figure(figsize= (18, 10))
plt.plot(year_count.index, year_count['Date'], marker= '')

plt.xlabel('Years')
plt.ylabel('Total Accidents')
plt.show()
```

```python
df['Year'] = df['DateTime'].dt.year
```

```python
plt.figure(figsize=(20, 10))
per_year_plot = sns.histplot(data = df, x = 'Year', bins = 200, color = "blue")
per_year_plot.set(title = "Crashes per year")
fig = per_year_plot.get_figure()
fig.savefig('per_year_plot.png')
```

```python
operator_count = df.groupby('Operator')[['Fatalities']].count()

top_operators = operator_count.sort_values(by='Fatalities', ascending=False).head(5)

plt.figure(figsize=(18, 10))
plt.plot(top_operators.index, top_operators['Fatalities'], marker='.')
plt.xlabel('Operator')
plt.ylabel('Total Fatalities (Top 10)')

plt.ylim(0, top_operators['Fatalities'].max() + 50)  

plt.show()
```
